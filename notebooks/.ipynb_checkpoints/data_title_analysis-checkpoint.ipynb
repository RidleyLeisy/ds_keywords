### Top level wrangling

```python
def clean_location(df):
    city = df['location'].str.split(',',expand=True)
    zips = city[1].str.split(' ',expand=True)
    zips.drop(0,axis=1,inplace=True)
    locs = (pd.concat([city[0],zips],axis=1)).iloc[:,0:3]
    locs.columns = ['city','state','zip']
    df = pd.concat([df,locs],axis=1).drop('location',axis=1)
    return df
```

```python
def remove_emp(x):
    """Remove equal employment"""
    return x.split('equal employment')[0]
```

```python
def remove_phrases(x):
    """Remove common application phrases"""
    return x.replace('sexual orientation', '').replace('race','').replace('color','').replace('equal opportunity','').replace('gender', '').replace('identity','').replace('-','')
```

```python
df = clean_location(df)
```

```python
# checking for nulls
print(df['description'].isnull().sum())
df = df.loc[~df['description'].isnull()]
```

```python
df['description'] = df['description'].str.lower()
df['position'] = df['position'].str.lower()
```

```python
df['description'] = df['description'].apply(remove_phrases)
df['description'] = df['description'].apply(remove_emp)
```

```python
# categorizing job titles
categories = {'data scientist I':0,
              'data':0,
              'analyst':0,
              'analytics':0,
              'scientist':0,
              'assistant':0,
              'data analyst I': 0,
              'intern':0,
              'junior':0,
              'data analyst': 0,
              'data scientist': 0,
              'machine learning': 1,
              'deep learning':1,
              'product manager': 1,
              'mid-level':1,
              'statistician':1,
              'statistical programmer':1,
              'senior': 1,
              'principal': 2,
              'manager': 2,
              'lead':2,
              'director':2,
              'vp':2}
```

```python
def categorize(x):
    group = "None"
    for key in categories:
        if key in x:
            group = categories[key]
            continue
    return group
```

## Word tokenizer

A few parts of speech and dependencies in text we will look at are...
- POS 
    - NOUN, ADJ
- DEP 
    - DOBJ, CONJ, AMOD, NUM, ADVMOD, QUANTMOD, MEASURE

For reference check out
https://nlp-ml.io/jg/software/pac/standep.html

```python
def nlp(series):
    nlp = English()
    nlp = spacy.load("en_core_web_sm",disable=['textcat'])
    corpus = [x for x in series]
    docs =[]
    for doc in nlp.pipe(corpus,batch_size=100,n_process=-1):
        docs.append(doc)
    return docs
```

```python
def nlp_filter(tokenized_docs):
    POS = ['ADJ','NOUN']
    DEP = ['DOBJ', 'CONJ', 'AMOD', 'NUM', 'ADVMOD', 'QUANTMOD', 'MEASURE']
    
    filtered_docs = []
    for doc in tokenized_docs:
        for word in doc:
            if ((word.pos_ in POS) | (word.dep_ in DEP)) & (word.is_stop == False):
                filtered_docs.append(word)
    return filtered_docs
```

```python
def create_count_plot(freq, title):
    
    x = [x[0] for x in freq]
    y = [x[1] for x in freq]
    fig, ax = pyplot.subplots(figsize=(10,7))

    sns.barplot(x,y,palette='BuGn_r',ax=ax)
    sns.despine()
    ax.set_xlabel('Words In Job Description',size=16)
    ax.set_ylabel('Freq (word count)',size=16)
    ax.set_title(title,size=20)
    for item in ax.get_xticklabels():
        item.set_rotation(45)
    
```

```python
def create_scatter_plot(df, title, hue):
    fig, ax = pyplot.subplots(figsize=(10,7))
    sns.scatterplot(df[0], df['perc'], hue=hue, data=df, s=50)
    sns.despine()
    ax.tick_params(labelsize='large')
    ax.set_xlabel('Words In Job Description',size=16)
    ax.set_ylabel('Freq (%)',size=16)
    ax.set_title(title,size=20)
    ax.legend(fontsize=12)
    for item in ax.get_xticklabels():
        item.set_rotation(90)
```

```python
def get_percentiles(dist):
    """Top ten most common words in percents"""
    summation = 0
    for num in dist:
        summation += num[1]
    perc = []
    for word in dist:
        perc.append(word[1] / summation)
    return perc
```
